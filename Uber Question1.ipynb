```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/Users/xiaoyan/Documents/oa/uber data analyst/green_tripdata_2014-06.csv')
df.VendorID = pd.to_datetime(df.VendorID)
print(df.shape)
```

```python
df = df.loc[~((df['Trip_distance'] == 0))]
print(df.shape)
```

```python
df["Trip_count"] = 1
df.head(5)
```

```python
dgroup1 = df.groupby(pd.Grouper(key='VendorID', freq='1D')).sum().sort_values(by='Passenger_count', ascending=False)
dgroup1.head(5)
```

```python
dgroup2 = df.groupby(pd.Grouper(key='VendorID', freq='1D')).sum().sort_values(by='Fare_amount', ascending=False)
dgroup2.head(5)
```

```python
print(dgroup1["Passenger_count"].max())
print(dgroup2["Fare_amount"].max())
```
